```python
import numpy as np
import matplotlib.pyplot as plt
```

### Уровень 0:

Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4\sin(2x) + \cos(3x+3)$

В точке x = 1.


$f'(x) = 5x^4 + 4\cos(2x)\cdot 2 - \sin(3x+3)\cdot 3 = 5x^4 + 8\cos(2x)- 3\sin(3x+3)$  
$f'(1) = 5\cdot 1^4 + 8\cos(2\cdot 1)- 3\sin(3 \cdot 1+3) = 5 + 8 \cos(2) - 3 \sin(6)$

```python
5 + 8 * np.cos(2) - 3* np.sin(6)
```

$f''(x) = 20x^3 -8\sin(2x)\cdot 2- 3\cos(3x+3)\cdot 3 = 20x^3 -16\sin(2x)- 9\cos(3x+3)$  
$f''(1) = 20\cdot 1^3 -16\sin(2\cdot 1)- 9\cos(3 \cdot 1+3)$  

```python
20 - 16 * np.sin(2) - 9* np.cos(6)
```

### Уровень 1:

Постройте график первой и второй производной функции:

$f(x) = \sin^5(2x+1)$

На отрезке [-5; 5].


$f'(x) = 5 \sin^4(2x+1) \cdot \cos(2x+1) \cdot 2 = 10 \sin^4(2x+1)\cos(2x+1)$  

$f''(x) = 10 \left( 4\sin^3(2x+1)\cos(2x+1)\cdot 2 \cdot \cos(2x+1)- \sin^4(2x+1) \sin(2x+1)\cdot 2 \right)  = 80\sin^3(2x+1)\cos^2(2x+1) - 20\sin^5(2x+1) $

```python
#График функции
y = lambda x: np.sin(2*x+1)**5
fig = plt.subplots()
fig[1].grid(color = "blue", linestyle=':', linewidth=0.3)
x = np.linspace(-5, 5, 1000)
plt.plot(x, y(x))
plt.show()
```

```python
#График первой производной
y = lambda x: 10*np.sin(2*x+1)**4*np.cos(2*x+1)
fig = plt.subplots()
fig[1].grid(color = "blue", linestyle=':', linewidth=0.3)
x = np.linspace(-5, 5, 1000)
plt.plot(x, y(x))
plt.show()
```

```python
#График второй производной
y = lambda x: 80*np.sin(2*x+1)**3*np.cos(2*x+1)**2 - 20*np.sin(2*x+1)**5
fig = plt.subplots()
fig[1].grid(color = "blue", linestyle=':', linewidth=0.3)
x = np.linspace(-5, 5, 1000)
plt.plot(x, y(x))
plt.show()
```

#### Задание 1:

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.


```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(-10, 10, 500)

ax.plot(x, sts.norm.pdf(x, loc = 0, scale = 1), lw=2, label='loc = 0, scale = 1')
ax.plot(x, sts.norm.pdf(x, loc = 3, scale = 2), lw=2, label='loc = 3, scale = 2')
ax.plot(x, sts.norm.pdf(x, loc = -3, scale = 2), lw=2, label='loc = -3, scale = 2')
ax.plot(x, sts.norm.pdf(x, loc = -5, scale = 0.5), lw=2, label='loc = -5, scale = 0.5')
ax.plot(x, sts.norm.pdf(x, loc = 5, scale = 0.5), lw=2, label='loc = 5, scale = 0.5')

plt.legend(loc = 'upper right')

plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(0, 10, 500)

ax.plot(x, sts.expon.pdf(x, loc = 0, scale = 1), lw=2, label='loc = 0, scale = 1')
ax.plot(x, sts.expon.pdf(x, loc = 0, scale = 2), lw=2, label='loc = 0, scale = 2')
ax.plot(x, sts.expon.pdf(x, loc = 0, scale = 3), lw=2, label='loc = 0, scale = 3')
ax.plot(x, sts.expon.pdf(x, loc = 0, scale = 5), lw=2, label='loc = 0, scale = 5')
ax.plot(x, sts.expon.pdf(x, loc = 0, scale = 10), lw=2, label='loc = 0, scale = 10')

plt.legend(loc = 'upper right')

plt.show()
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(-20, 20, 500)

ax.plot(x, sts.t.pdf(x, df=1, loc = 0, scale = 1), lw=2, label='df = 1, loc = 0, scale = 1')
ax.plot(x, sts.t.pdf(x, df=100, loc = 0, scale = 1), lw=2, label='df = 100, loc = 0, scale = 1')
ax.plot(x, sts.t.pdf(x, df=20, loc = -3, scale = 3), lw=2, label='df = 20, loc = -3, scale = 3')
ax.plot(x, sts.t.pdf(x, df=50, loc = 5, scale = 5), lw=2, label='df = 50, loc = 5, scale = 5')
ax.plot(x, sts.t.pdf(x, df=100, loc = -5, scale = 10), lw=2, label='df = 100, loc = -5, scale = 10')

plt.legend(loc = 'upper right')

plt.show()
```

Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$, а также вероятность попадания в отрезок $[3.5;7]$, т.е. P(3.5 < $\xi$ < 7).


```python
a = 0
b = 5

x = np.linspace(-1, 6, 1000)
pdf = sts.uniform(a, b-a).pdf(x)
plt.plot(x, pdf)
plt.show()

# C = 0.2
```

```python
pdf.mean()
```

```python
sts.uniform(a, b-a).cdf(7) - sts.uniform(a, b-a).cdf(3.5)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
gp_apps_df = pd.read_csv('data/googleplaystore.csv')
gp_apps_df.head()
```

```python
gp_apps_df.info()
```

```python
for i, install in enumerate(gp_apps_df.Installs):
    gp_apps_df.loc[i, 'Installs'] = gp_apps_df.loc[i, 'Installs'].replace('+','')
    gp_apps_df.loc[i, 'Installs'] = gp_apps_df.loc[i, 'Installs'].replace(',','')
gp_apps_df.Installs = gp_apps_df.Installs.map(int)
```

```python
top_categories = gp_apps_df.Category.value_counts().sort_values(ascending = False).head(10).index.values

sns.boxplot(x="Category", y="Rating",
            data=gp_apps_df[gp_apps_df.Category.isin(top_categories)])
```

**Вывод:** Среди популярных категорий наиболее высокие оценки пользователей получают приложения категории PERSONALIZATION, тогда как наиболее низкие - приложения категории LIFESTYLE.

При этом приложения категории PERSONALIZATION ни разу не получали оценок ниже 2.5 (что в данном случае является естественным выбросом)

Также стоит отметить, что несмотря на то, что приложения категории LIFESTYLE в среднем оцениваются довольно низко, для них отсутствуют оценки ниже 1.5. Тогда как, скажем, для категории TOOLS, также имуюший довольно длинный нижний "хвост", встречаются приложения с рейтингом 1.

```python
gp_apps_df.plot(kind='scatter', 
        x='Rating', 
        y='Reviews', 
        title='Correlation between Ratings and Reviews')
plt.show()
```

**Вывод:** Пользователи склонны чаще оставлять комментарии к приложениям, которые они оценивают выше. Однако это также может быть связано с агрессивной политикой приложения в области сбора оценок и комментариев.

```python
rating_by_installs_df = gp_apps_df.groupby('Installs').agg({'Rating' : 'mean'}).iloc[1:,:]
```

```python
rating_by_installs_df.plot(kind='bar', 
                          title='Installs by rating',
                          xlabel='Installs',
                          ylabel='Rating', rot=30, ylim=(3.5, 5.5))
```

**Вывод:** Приложения с меньшим количеством установок в среднем имеют более высокие оценки. 

Тем не менее это может быть связано с тем, что для небольших приложений проще собрать положительные отзывы, используя ближайшие круги общения разработчиков или же "накручивая" положительные отзывы. Для относительно больших приложений (от 5000 установок) количество объективных оценок пользоваталей начинает иметь больший вес, а потому средний рейтинг становится ниже. 

При этом для каждой следующей после 5000 установок категории средний рейтинг становится выше. Это может говорить о более высоком качестве приложений, имеющих более высокие рейтинги.

Приложения, имеющие более 100 млн. установок являются лидерами рынка в своих сферах. Для них наблюдается обратная корреляция среднего рейтинга и количества установок. Это может быть связано, во-первых, с тем, что многие пользователи устанавливают такие приложения в силу тренда , вне зависимости от того, насколько они им нравятся и, во-вторых, с консолидированной деятельностью конкурентов.

```python
intalls_by_content_rating = gp_apps_df.pivot_table(index='Installs', columns='Content Rating', values='App', aggfunc='count', fill_value=0)
```

```python
intalls_by_content_rating.plot(kind='bar', stacked=True, title="Apps ammount by Installs and Content rating",
                          xlabel='Installs',
                          ylabel='Amount')
```

**Вывод:** Можно заметить, что график имеют довольно необычную форму: в большинстве столбцов, начинающихся с "1" (10, 100, 1000, ...) значительно больше приложений, чем в столбцах, начинающихся с "5" (50, 500, 5000, ...), хотя, казалось бы, значения должны измениться постепенно от столбца к столбцу. Это может говорить о некой специфике ведения/сбора данных.

Также отметим, что большинство приложений имеют 100 000+ установок. Это, однако, не говорит о том, что для любого выпускаемого приложения шанс набрать именно 100 000 установок выше, чем другое число установок. Это связано с тем, что приложения, имеющие мало установок попросту "не выживают" и через некоторое время удаляются из маркетплейса.

Наконец, очевидно, что в разразе любого количества установок наибольшую часть приложений составляют приложения, предназначенные для любых пользоваталей (Content Rating = Everyone). Сразу после них идут приложения для детей (Content Rating = Teen)
